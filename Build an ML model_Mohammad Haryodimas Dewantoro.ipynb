## Metode
Berikut adalah langkah-langkah untuk metode "Build an ML model":

- Pemilihan Model: Pilihlah model yang terbaik berdasarkan hasil evaluasi.


Model ML yang dipilih terdiri atas:
## 1. Decision Tree
Decision Tree adalah teknik Supervised learning yang dapat digunakan untuk masalah klasifikasi dan Regresi, namun sebagian besar lebih disukai untuk menyelesaikan masalah Klasifikasi. Ini adalah pengklasifikasi terstruktur pohon, di mana node internal mewakili fitur kumpulan data, cabang mewakili aturan keputusan, dan setiap node daun mewakili hasilnya.
Dengan membagi data secara rekursif menjadi himpunan bagian yang lebih kecil hingga setiap himpunan bagian homogen sehubungan dengan variabel tujuan, maka pohon keputusan akan dibangun. Karakteristik paling signifikan yang secara akurat memprediksi variabel target menjadi dasar pemisahan.
## 2. Random Forest
Random Forest (RF) adalah algoritma ensemble learning yang didasarkan pada metode 'mengantongi(bagging)' pohon. Pohon-pohon tersebut dibuat secara independen menggunakan sampel dari kumpulan data, dan suara terbanyak diambil untuk prediksinya. Yang membedakan RF adalah pengenalan keacakan pada metode pembuatan pohon klasifikasi dan regresi. Artinya setiap node dibagi berdasarkan prediktor terbaik dari subkumpulan pohon yang dipilih secara acak. Algoritma ini mengambil sampel bootstrap dari data, menumbuhkan pohon klasifikasi atau regresi, dan kemudian memprediksi data baru dengan menggabungkan prediksi dari hutan pohon yang dikembangkan.
## 3. Gradient Boosting
Gradient Boosting adalah algoritme ensemble learning yang membangun serangkaian pembelajar yang lemah (biasanya pohon keputusan) secara berurutan, dengan setiap pembelajar berikutnya mengoreksi kesalahan pembelajar sebelumnya. Ini meminimalkan fungsi kerugian dengan menambahkan model baru yang memprediksi sisa atau kesalahan yang dibuat oleh model yang sudah ada. Gradient Boosting memberikan kinerja yang lebih baik dibandingkan dengan algoritma peningkatan tradisional dengan menggabungkan teknik regularisasi dan dikenal karena efisiensi dan efektivitasnya dalam tugas klasifikasi dan regresi.
## 4. Support Vector Machines
Support Vector Machine (SVM) adalah algoritma supervised learning yang terutama digunakan untuk tugas klasifikasi. Hal ini bertujuan untuk menemukan hyperplane yang paling baik memisahkan kelas-kelas dalam ruang fitur. SVM beroperasi dengan memetakan data masukan ke dalam ruang fitur berdimensi tinggi dan kemudian menentukan hyperplane optimal yang memisahkan kelas-kelas dengan margin terbesar.

## Clean Data | Membersihkan Data

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
duplicate_rows = data.duplicated()

# Count the number of duplicate rows
num_duplicates = duplicate_rows.sum()
print("Number of Duplicate Rows:", num_duplicates)

# Display duplicate rows
duplicate_data = data[duplicate_rows]
print("Duplicate Rows:")
print(duplicate_data)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical column
label_encoder = LabelEncoder()
data['Sex'] = label_encoder.fit_transform(data['Sex'])
data['BP'] = label_encoder.fit_transform(data['BP'])
data['Cholesterol'] = label_encoder.fit_transform(data['Cholesterol'])
data
```

```python
x = data.drop(['Drug'], axis=1)
y = data['Drug']
from sklearn.model_selection import train_test_split

# Split the data into training (80%) and testing (20%) sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Training Model ML

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Build and train Decision Tree Classifier
clf_decision_tree = DecisionTreeClassifier(random_state=42)
clf_decision_tree.fit(x_train, y_train)
y_pred_decision_tree = clf_decision_tree.predict(x_test)
score_decision_tree = clf_decision_tree.score(x_test,y_test)
print("Decision Tree Classifier Score:", score_decision_tree)
```

```python
# Build and train Random Forest Classifier
clf_random_forest = RandomForestClassifier(random_state=42)
clf_random_forest.fit(x_train, y_train)
y_pred_random_forest = clf_random_forest.predict(x_test)
score_random_forest = clf_random_forest.score(x_test,y_test)
print("Random Forest Classifier Score:", score_random_forest)
```

```python
# Build and train Gradient Boosting Classifier
clf_gradient_boosting = GradientBoostingClassifier(random_state=42)
clf_gradient_boosting.fit(x_train, y_train)
y_pred_gradient_boosting = clf_gradient_boosting.predict(x_test)
score_gradient_boosting = clf_gradient_boosting.score(x_test,y_test)
print("Gradient Boosting Classifier score:", score_gradient_boosting)
```

```python
# Build and train Support Vector Machine (SVM) Classifier
clf_svm = SVC(random_state=42)
clf_svm.fit(x_train, y_train)
y_pred_svm = clf_svm.predict(x_test)
score_svm = clf_svm.score(x_test,y_test)
print("SVM Classifier score:", score_svm)
```

Dari 4 hasil akurasi model tersebut, kita ambil yang paling akurat yaitu Decision Tree, Random Forest, dan Gradient Boosting. Namun, Ketiga model perlu dilakukan Cross-Validation & Hyperparameter Tuning.

```python
#cross-validation
from sklearn.model_selection import StratifiedShuffleSplit, cross_val_score

cvs = StratifiedShuffleSplit(n_splits = 4, test_size = 0.2, random_state = 42)

score_dt = cross_val_score(clf_decision_tree, x, y, cv = cvs)
score_rf = cross_val_score(clf_random_forest, x, y, cv = cvs)
score_gd = cross_val_score(clf_gradient_boosting, x, y, cv = cvs)

print(score_dt)
print(score_rf)
print(score_gd)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search_dt = GridSearchCV(estimator = clf_decision_tree, param_grid = param_grid_dt) 
grid_search_dt.fit(x_train, y_train) 
```

```python
grid_search_rf = GridSearchCV(estimator = clf_random_forest, param_grid = param_grid_rf) 
grid_search_rf.fit(x_train, y_train) 
```

```python
grid_search_gd = GridSearchCV(estimator = clf_gradient_boosting, param_grid = param_grid_gd) 
grid_search_gd.fit(x_train, y_train) 
```

Hasil dari GridSearchCV belum tentu yang paling optimal, jadi saya juga melakukan tuning secara manual.

```python
# Decision tree model with hyperparameter tuning
model_dt = DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_leaf=2, random_state=42) 
model_dt.fit(x_train, y_train)
y_pred_model_dt = model_dt.predict(x_test)

score_model_dt = model_dt.score(x_test,y_test)
print("Decision Tree Classifier score:", score_model_dt)
```

```python
# Random Forest Classifier model with hyperparameter tuning
model_rf = RandomForestClassifier(max_depth= 6, max_leaf_nodes= 6, n_estimators= 25, random_state=42)
model_rf.fit(x_train, y_train)
y_pred_model_rf = model_rf.predict(x_test)

score_model_rf = model_rf.score(x_test,y_test)
print("Random Forest Classifier score:", score_model_rf)
```

```python
# Gradient Boosting model with hyperparameter tuning
model_gd = GradientBoostingClassifier(learning_rate=0.01, n_estimators=200, random_state=42)
model_gd.fit(x_train, y_train)
y_pred_model_gd = model_gd.predict(x_test)

score_model_gd = model_gd.score(x_test,y_test)
print("Gradient Boosting Classifier score:", score_model_gd)
```

## Model Selection | Pemilihan Model

Setelah evaluasi masing-masing model, model yang dipilih yaitu model yang akurasinya tinggi, yaitu Decision Tree, Random Forest, dan Gradient Boosting

```python
# Create a DataFrame with the actual and predicted values
results_df = pd.DataFrame({'Actual': y_test, 'Predicted DT': y_pred_model_dt, 'Predicted RF': y_pred_model_rf, 'Predicted GD': y_pred_model_gd })
print(results_df)
```

```python
# Assuming results_df is the DataFrame with 'Actual' and 'Predicted' columns
# Calculate the number of correct predictions
correct_predictions_dt = (results_df['Actual'] == results_df['Predicted DT']).sum()
correct_predictions_rf = (results_df['Actual'] == results_df['Predicted RF']).sum()
correct_predictions_gd = (results_df['Actual'] == results_df['Predicted GD']).sum()


# Calculate the total number of predictions
total_predictions = len(results_df)

# Calculate the percentage of accurate predictions
accuracy_percentage_dt = (correct_predictions_dt / total_predictions) * 100
accuracy_percentage_rf = (correct_predictions_rf / total_predictions) * 100
accuracy_percentage_gd = (correct_predictions_gd / total_predictions) * 100

print("Accuracy Percentage of Decision Tree:", accuracy_percentage_dt)
print("Accuracy Percentage of Random Forest :", accuracy_percentage_rf)
print("Accuracy Percentage of Gradient Boosting:", accuracy_percentage_gd)
```
